```python
def create_employee(num_employee):
  employee_list =[]

  for i in range(num_employee):
    employee={}
    employee['First_Name'] = fake.first_name()
    employee['last_Name'] = fake.last_name()
    employee['Job'] = fake.job()
    employee['Department'] = fake.random_element(elements=('HR','Admin', 'IT','Marketing', 'Finance'))
    employee['Role'] = fake.random_element(elements=('Developer', 'Manager', 'Analyst','Associate'))
    employee['Email'] = fake.email()
    employee['Phone'] = fake.phone_number()
    employee['Address'] = fake.address()
    employee['City'] = fake.city()
    employee['State'] = fake.state()
    employee_list.append(employee)

  return pd.DataFrame(employee_list)
```

```python
data = create_employee(5000)
```
